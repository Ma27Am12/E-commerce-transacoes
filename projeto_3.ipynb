## Limpeza e transformação dos dados

```python
df.info()
```

Verificando se há duplicatas


```python
duplicatas = df.duplicated()
print(duplicatas.sum())
```

verificando se há valores nulos

```python
valores_nulos = df.isnull().sum()
print(valores_nulos)
```

modificando o Purchase_Date de d-m-y para d/m/y

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'],format='%d-%m-%Y').dt.strftime('%d/%m/%Y')
```

```python
df.head()
```

```python
count_payment_method = df['Payment_Method'].value_counts()
```

```python
df['Count_Payment_Method'] = df['Payment_Method'].map(count_payment_method)
```
